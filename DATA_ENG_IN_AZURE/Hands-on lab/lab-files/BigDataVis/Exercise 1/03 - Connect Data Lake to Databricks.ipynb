### Step 1: Set the storage account connection details & location

To get started, we need to set the connection details of the Azure storage account, and the raw data location:

```
storage_account_name = "<Your Storage Account Name>"
storage_account_access_key = "<Your Storage Access Key>"
storage_container_name = "sparkcontainer"
data_location_folder = "Triage"
```

### Step 2: Mount the storage account to the Databricks cluster

To get started, we need to set the location of the raw data files.

```
def mount_storgae(storage_account, access_key, container_name, folder):
    dbutils.fs.mount(
        source = "wasbs://{0}@{1}.blob.core.windows.net".format(container_name, storage_account),
        mount_point = "/mnt/{0}".format(container_name),
        extra_configs = {"fs.azure.account.key.{0}.blob.core.windows.net".format(storage_account): access_key})
```

```
mount_storgae(storage_account_name, storage_account_access_key, storage_container_name, data_location_folder)
```
